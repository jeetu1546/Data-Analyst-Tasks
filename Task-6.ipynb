```python
#Crosstabulation to count the occurance of each director
director_counts = pd.crosstab(index=df['director'], columns='count')

# Sorting the directors based on the count of their content
director_counts_sorted = director_counts.sort_values(by='count', ascending=False)

#Getting the top three directors
top_three_directors= director_counts_sorted.head(3)

print("Top three directors with the most content:")
print(top_three_directors)
```

```python
#Let's find the answer of our 2nd question.
#2.who are the top three directors with the most content in the dataset?
```

```python
#Crosstabulation to count the occurance of each director
director_counts = pd.crosstab(index=df['director'], columns='count')

# Sorting the directors based on the count of their content
director_counts_sorted = director_counts.sort_values(by='count', ascending=False)

#Getting the top three directors
top_three_directors= director_counts_sorted.head(3)

print("Top three directors with the most content:")
print(top_three_directors)
```

```python
#Getting the top three directors
top_three_directors = director_counts_sorted.head(3)
# Plotting the top three directors
plt.figure(figsize=(8, 4))
sns.barplot(x=top_three_directors.index, y='count', data=top_three_directors)
plt.xlabel('Director')
plt.ylabel('Number of Titles')
plt.title('Top Three Directors with Most Content')
plt.show()
```

```python
#Let's move to question 3
#3.Which country has produced the highest number of movies and TV shows combined in the dataset ?
```

```python
# Assuming 'df' is your DataFrame containing the dataset
# Replace 'df' with the name of your DataFrame if different

# Grouping the data by 'country' and summing the counts of movies and TV shows
country_counts = df['country'].value_counts()

# Getting the country with the highest count
top_country = country_counts.idxmax()
highest_count = country_counts.max()

print("Country with the highest number of movies and TV shows combined:")
print("Country:", top_country)
print("Number of titles:", highest_count)
```

```python
# Grouping the data by 'country' and summing the counts of movies and TV shows
country_counts = df['country'].value_counts()

#Getting the top 10 countries with the highest number of titles
top_countries = country_counts.head(5)

#plotting the bar chart
plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.title('Top 5 Countries with the Highest Number of Movies and TV Shows combined')
plt.tight_layout()
plt.show()
```
